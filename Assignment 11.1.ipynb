```python
import numpy as np
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup
import urllib.request
import nltk
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
soup
```

```python
len(soup.find_all("p"))       #finding how many p tags above data contains
```

```python
text = soup.get_text(strip=True)             #eliminating all HTML tags
text
```

now we have eliminated all the html tags from the text and converted text into string.<br>
now spliting above text into list where splitting will take place on 'whitspace'

```python
text_v1 = text.split()
text_v1
```

In the above text there are many non alphabatic words present. so eliminating non alphabatic words.<br>
also converting each alphabatic term into lowercase form. this will help us to find same words with different case.<br>
e.g. 'the' and 'The'

```python
text_v2 = [word.lower() for word in text_v1 if word.isalpha()]
len(text_v2)
```

```python
text_v2
```

now finding how the different words are distributed along the text? what is their freq of occurance?

```python
fdist1 = nltk.FreqDist(text_v2)
fdist1
```

```python
fdist1.plot(50, cumulative=True)
```

we can see there are repeatition of many words. to count only the number of unique words.
